This is documentation on the exploration of making a website under the constraints:
* Free - Everything should be without any cost.
* Easy - The construction of the site should be easy, without the need to learn HTML.
-------------------------------------------------
Each constraint add a higher complexity. lets explor them:
* With charge(Not free) - There are sites that offer "all included", but you need to pay for that.
* Free and not easy - static site that are free to use, but you need
If you ever want to make your own site?

-------------------------------------------------
First of all we need a free host for the website.
Github & Bitbucket offer a free hosting a static website[1], just creat a new repository with ending of "<workspaceid>.bitbucket.io" and set the main branch to name "gh-pages".
but we also want to create a website asily**.

-------------------------------------------------
For creating websites easily we need content management system(CMP), which is much easier than write code(e.g. HTML) or other static site generators(e.g. Jekyll).
Currently one of the best tools for that is "WordPress", CMP tool that 20%-30% of the websites are written with it![2]

-------------------------------------------------
We want to install WordPress locally, but if we install only "WordPress", we will need to install couple staff in addition.
The easiest way is to install "Local", manage everything so we can focus only on the creation of the website.

-------------------------------------------------
After installing it we want a tool that will help us to metigate the gap between the site build in the CMP anf the static site that the Github/Bitbucket needs.
After the installsion of the "Local" we need to add a plugin that converts the CMP to static site - the plugin name is "Simply Static"[4].

-------------------------------------------------
the last thing we want an editor for make thins easy to edit, the default editor is Gutenberg, much easier editor is Elementor.
I recommand Tyler Moore, which gives clear explantions, check [5] for example(you can skip the first part and go stright to the explanation)
